```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Image path
img_path = "Images/blox.jpg"

# Read image as colored
img_color = cv.imread(img_path)
```

```python
# Convert the image to grayscale, and show it
img = cv.cvtColor(img_color, cv.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# We use the algorithm to search corners for the top 25 who are at least 10 pixels
# each other
corners = cv.goodFeaturesToTrack(img, 25, 0.01, 10)
corners = np.intp(corners)
```

```python
# We iter per each corner that was get by the model
for i in corners:
    # Obtain the coordinates of the corner
    x, y = i.ravel()

    # Draw over the image a green circle in the extracted coordinates
    cv.circle(img_color, (x,y), 3, (0,255,0),-1)
```

```python
# show on screen the image with the top 25 corners
plt.imshow(img_color)
plt.show()
```
