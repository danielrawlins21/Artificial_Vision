```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Image Path
path_to_image = 'Image/home.jpg'
```

```python
# Open the image with openCV in gray scale(0 option)
img = cv.imread(path_to_image, 0)
```

```python
# Showing the image
plt.imshow(img, cmap="gray")
plt.show()
```

```python
# Calulate the histogram with frequencies from 0 to 255
plt.hist(img.ravel(), 255,[0,255])
plt.show()
```

```python
# Opening the same image
img1 = cv.imread(path_to_image)
```

```python
# Convert the image to RGB
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
plt.imshow(img1)
plt.show()
```

```python
channels={
    0:'red',
    1:'green',
    2:'blue'
}
```

```python
# Plotting the histogram for every channel
for channel, color in channels.items():
    histr = cv.calcHist([img1],[channel],None,[256],[0,256])
    plt.plot(histr, color=color)
    plt.xlim([0,255])
plt.show()
```
