```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Adding image path
img_path = "../Deteccion de esquinas/Images/blox.jpg"
```

```python
# opening image
img = cv.imread(img_path, 0)
```

```python
# Creating an ORB object
orb = cv.ORB_create()
```

```python
# Call the method to detect the features and save the result called "features"
features = orb.detect(img)
```

```python
# Describe our features and save into a variable called "description"
features, description = orb.compute(img, features)
```

```python
# Draw the features over the image to color, and show it
img2 = cv.drawKeypoints(img, features, None, color=(0,255,255))
plt.imshow(img2)
plt.show()
```
