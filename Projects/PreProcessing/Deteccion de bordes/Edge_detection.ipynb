```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from skimage import img_as_ubyte
```

```python
# Opening "astronaut" example image
img = img_as_ubyte(data.astronaut())
plt.imshow(img)
plt.show()
```

```python
# Applying a gaussian filter as we studied in the section 2.5 to remove noise and fix the edges
img_blur = cv.GaussianBlur(src=img, ksize=(3,3), sigmaX=0)

# Show the image in screen
plt.imshow(img_blur)
plt.show()
```

```python
# Convert the image to grayscale for edge search and show in screen
img_blur_gray = cv.cvtColor(img_blur, cv.COLOR_BGR2GRAY)
plt.imshow(img_blur_gray, cmap="gray")
plt.show()
```

# Sobel Model

```python
# Edge detection with sobel, into x and y axis, with a (5 x 5) kernel
sobelxy = cv.Sobel(src=img_blur_gray,ddepth=cv.CV_8U, dx = 1, dy = 1, ksize=5)

# Show image on screen
plt.imshow(sobelxy, cmap="gray")
plt.show()

# Edge detection with sobel, into x and y axis, with a (3 x 3) kernel
sobelxy1 = cv.Sobel(src=img_blur_gray,ddepth=cv.CV_16U, dx = 1, dy = 1, ksize=9)

# Show image on screen
plt.imshow(sobelxy1, cmap="gray")
plt.show()
```

# Laplacian Model

```python
# Edge detection with laplacian model with a ( 5 x 5 ) kernel
laplacian = cv.Laplacian(src=img_blur_gray, ddepth=cv.CV_8U, ksize=5)

# Show image on screen
plt.imshow(laplacian, cmap="gray")
plt.show()
```

```python
# Edge detection with laplacian model with a ( 3 x 3 ) kernel
laplacian1 = cv.Laplacian(src=img_blur_gray, ddepth=cv.CV_8U, ksize=3)

# Show image on screen
plt.imshow(laplacian1, cmap="gray")
plt.show()
```

```python
# Edge detection with laplacian model with a ( 7 x 7 ) kernel
laplacian2 = cv.Laplacian(src=img_blur_gray, ddepth=cv.CV_16U, ksize=5)

# Show image on screen
plt.imshow(laplacian2, cmap="gray")
plt.show()
```

# Canny Model

```python
# Edge detection with Canny model

canny = cv.Canny(image=img_blur_gray, threshold1=100, threshold2=200)

# Show image on screen
plt.imshow(canny, cmap="gray")
plt.show()
```

```python
# Edge detection with Canny model

canny1 = cv.Canny(image=img_blur_gray, threshold1=50, threshold2=300)

# Show image on screen
plt.imshow(canny1, cmap="gray")
plt.show()
```

```python
# We detect the outline over the result of Canny model without Approach
contours, _ = cv.findContours(image=canny,
                              mode= cv.RETR_TREE,
                              method=cv.CHAIN_APPROX_NONE)
```

```python
# we draw a copy of the outlines from the original
img_copy = img.copy()
```

```python
# With the funtion we are going to draw outline to the color image, with the gross and kind of line
cv.drawContours(image=img_copy,
                contours=contours,
                contourIdx=-1,
                color=(0,255,0),
                thickness=2,
                lineType=cv.LINE_AA)
```

```python
# Show image
plt.imshow(img_copy)
plt.show()
```

```python
# We detect the outline over the result of Canny model with simple Approach
contours1, _1 = cv.findContours(image=canny,
                              mode= cv.RETR_TREE,
                              method=cv.CHAIN_APPROX_SIMPLE)
```

```python
# Copy the original image
img_copy1 = img.copy()
```

```python
# Repeat the process but with a simple approach
cv.drawContours(image=img_copy1,
                contours=contours1,
                contourIdx=-1,
                color=(0,255,0),
                thickness=2,
                lineType=cv.LINE_AA)
```

```python
plt.imshow(img_copy1)
plt.show()
```
