```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fill the relative path of the image
path_to_image = 'Image/roya.jpg'
#path_to_image = 'Image/oidio.jpg'
```

```python
# Read the image
img = cv.imread(path_to_image)
```

```python
# OpenCV use by default the space color RGB, but we have to change it to see the image
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
# We show the image
plt.imshow(img)
plt.show()
```

```python
# We transform the image to HSV
hsv = cv.cvtColor(img, cv.COLOR_RGB2HSV)
```

```python
# We define a range of orange to detect the Roya
lower_blue = np.array([10,90,110])
upper_blue = np.array([18,255,255])
```

```python
# We extract the pixels that their color is in our range
mask = cv.inRange(hsv, lower_blue, upper_blue)
```

```python
plt.imshow(mask)
plt.show()
```

```python
# select the pixels that we are found in our original image
res = cv.bitwise_and(img,img, mask=mask)
```

```python
# showing the results
plt.imshow(res)
plt.show()
```
