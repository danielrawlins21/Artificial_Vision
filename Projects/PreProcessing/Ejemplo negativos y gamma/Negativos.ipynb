```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# add the route to the image
path_to_image = '../Ejemplo histograma mono/Image/home.jpg'
```

```python
# Open the image in grayscale
img = cv.imread(path_to_image, 0)
```

```python
# Calculate the negative of the image
img_not = cv.bitwise_not(img)
```

```python
# Show the two images
plt.imshow(img, cmap="gray")
plt.show()
plt.imshow(img_not, cmap="gray")
plt.show()
```

```python
# Create gamma variable
gamma = 0.5
```

```python
# read again the image and convert it to RGB
color_img = cv.imread(path_to_image)
color_img = cv.cvtColor(color_img, cv.COLOR_BGR2RGB)
```

```python
# Create a search table to apply gamma correction
lookUpTable = np.empty((1,256), np.uint8)
for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)
    
```

```python
# Applying gamma correction to our search table
res = cv.LUT(color_img, lookUpTable)
```

```python
# Showing both images
plt.imshow(color_img)
plt.show()
plt.imshow(res)
plt.show()
```
